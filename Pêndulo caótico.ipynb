```python
import matplotlib.pyplot as plt
import numpy as np
import math 
```

```python
#cte globais
w = 0.7                   #Frequência angular
periodo = (2*np.pi)/w     #Período
c = 0.05                
x0 = 1                    #posição inicial
v0 = 0                    # velocidade inicial
dt = 0.01                 # passo de tempo
tempomax = 20*np.pi
tempoin = 800             #corta t.t.
```

```python
def v(v):       #velocidade angular           
    return v

def fa(t,F,x,v):   #Pêndulo equação
    return  -v*c -np.sin(x) + F*np.cos(w*t)
```

```python

def  RG4a(F,f,dt,tdi, tmax): 
    tf= []
    xf= []
    vf= []
    x = x0
    v = v0 
    i = 0
    t = 0
    while t < tdi:  
        k1  = v
        k1a = f(t,F,x,v)
        
        k2  = v + k1a*(dt/2)
        k2a = f(t + (dt/2), F, x + k1*(dt/2), v + k1a*(dt/2))
        
        k3 = v + k2a*(dt/2)
        k3a = f(t + (dt/2), F, x + k2*(dt/2), v + k2a*(dt/2))
        
        k4 = v + k3a*dt
        k4a = f(t + dt, F, x + k3*dt, v + k3a*dt)
        
        x = x + (dt/6)*(k1+ 2*k2 + 2* k3 + k4)
        v = v + (dt/6)*(k1a + 2*k2a + 2*k3a + k4a)
        
        i = i + 1
        t = i * dt
          
    i = 0 
    
    while t <= (tdi + tmax): 
        k1  = v
        k1a = f(t,F,x,v)
        
        k2  = v + k1a*(dt/2)
        k2a = f(t + (dt/2), F, x + k1*(dt/2), v + k1a*(dt/2))
        
        k3 = v + k2a*(dt/2)
        k3a = f(t + (dt/2), F, x + k2*(dt/2), v + k2a*(dt/2))
        
        k4 = v + k3a*dt
        k4a = f(t + dt, F, x + k3*dt, v + k3a*dt)
        
        x = x + (dt/6)*(k1+ 2*k2 + 2* k3 + k4)
        v = v + (dt/6)*(k1a + 2*k2a + 2*k3a + k4a)


        i = i + 1
        t = i * dt + tdi

        tf.append(t)
        xf.append(x)
        vf.append(v)

       
    return tf,xf,vf 
              
```

```python
def  RG4( F,f, dt, tdi, tmax): 
    tf= []
    xf= []
    vf= []
    x = x0
    v = v0 
    i = 0 
    t = 0
    while t < tdi: 
        k1  = v
        k1a = f(t,F,x,v)
        
        k2  = v + k1a*(dt/2)
        k2a = f(t + (dt/2), F, x + k1*(dt/2), v + k1a*(dt/2))
        
        k3 = v + k2a*(dt/2)
        k3a = f(t + (dt/2), F, x + k2*(dt/2), v + k2a*(dt/2))
        
        k4 = v + k3a*dt
        k4a = f(t + dt, F, x + k3*dt, v + k3a*dt)
        
        x = x + (dt/6)*(k1+ 2*k2 + 2*k3 + k4)
        v = v + (dt/6)*(k1a + 2*k2a + 2*k3a + k4a)

        i = i + 1
        t = i * dt

        
    i = 0 
    
    while t <= (tdi + tmax): 
        k1  = v
        k1a = f(t,F,x,v)
        
        k2  = v + k1a*(dt/2)
        k2a = f(t + (dt/2), F, x + k1*(dt/2), v + k1a*(dt/2))
        
        k3 = v + k2a*(dt/2)
        k3a = f(t + (dt/2), F, x + k2*(dt/2), v + k2a*(dt/2))
        
        k4 = v + k3a*dt
        k4a = f(t + dt, F, x + k3*dt, v + k3a*dt)
        
        x = x + (dt/6)*(k1+ 2*k2 + 2* k3 + k4)
        v = v + (dt/6)*(k1a + 2*k2a + 2*k3a + k4a)


        i = i + 1
        t = i * dt + tdi

        if x > math.pi:
            x -= 2 * math.pi
        elif x <= -math.pi:
            x += 2 * math.pi

        tf.append(t)
        xf.append(x)
        vf.append(v)

       
    return tf,xf,vf 
```

```python
def PC(F, f, dt, tdi, tmax): 
    tf= []
    xf= []
    vf= []
    x = x0
    v = v0
    i = 0 
    t = 0
    if t < tdi: 
        while t < tdi: 
            k1  = v
            k1a = f(t,F,x,v)
        
            k2  = v + k1a*(dt/2)
            k2a = f(t + (dt/2), F, x + k1*(dt/2), v + k1a*(dt/2))
        
            k3 = v + k2a*(dt/2)
            k3a = f(t + (dt/2), F, x + k2*(dt/2), v + k2a*(dt/2))
        
            k4 = v + k3a*dt
            k4a = f(t + dt, F, x + k3*dt, v + k3a*dt)
        
            x = x + (dt/6)*(k1+ 2*k2 + 2* k3 + k4)
            v = v + (dt/6)*(k1a + 2*k2a + 2*k3a + k4a)

            i = i + 1
            t = i * dt

    
    i = 0 
    
    while t <= (tdi + tmax):
        k1  = v
        k1a = f(t,F,x,v)
        
        k2  = v + k1a*(dt/2)
        k2a = f(t + (dt/2), F, x + k1*(dt/2), v + k1a*(dt/2))
        
        k3 = v + k2a*(dt/2)
        k3a = f(t + (dt/2), F, x + k2*(dt/2), v + k2a*(dt/2))
        
        k4 = v + k3a*dt
        k4a = f(t + dt, F, x + k3*dt, v + k3a*dt)
        
        x = x + (dt/6)*(k1+ 2*k2 + 2* k3 + k4)
        v = v + (dt/6)*(k1a + 2*k2a + 2*k3a + k4a)

        i = i + 1
        t = i * dt + tdi  

        if x > math.pi:
            x -= 2 * math.pi
        elif x <= -math.pi:
            x += 2 * math.pi

        if abs((t - tdi) % (periodo)) < (0.01):      #pegar cada perídodo do torque
            tf.append(t)
            xf.append(x)
            vf.append(v)
            

       
    return tf,xf,vf
```

```python
t4a,x4a,v4a = RG4a(0.4,fa, dt,tempoin,tempomax)
t4,x4,v4 = RG4( 0.4,fa, dt, tempoin, 50*tempomax)
t4b,x4b,v4b = PC(0.4, fa, dt, tempoin, 50*tempomax)
```

```python
fig = plt.figure(figsize=(30, 25))

tn4=[]
for i in t4a:
    tn4.append(i - tempoin)
    
plt.subplot(331)
plt.title("Velocidade angular por tempo com F = 0.4")
plt.scatter(tn4, v4a)
plt.xlim(0, tempomax)
plt.ylim(-4,4)
plt.ylabel("v")
plt.xlabel("t")
plt.grid()

plt.subplot(332)
plt.title("Phase-space plot F = 0.4")
plt.scatter(x4, v4)
plt.ylim(-4,4)
plt.xlim(-np.pi, np.pi)
plt.ylabel("v")
plt.xlabel("x")
plt.grid()

plt.subplot(333)
plt.title("Poincaré section F = 0.4")
plt.scatter(x4b, v4b)
plt.ylim(-4,4)
plt.xlim(-np.pi, np.pi)
plt.ylabel("v")
plt.xlabel("x")
plt.grid()
plt.savefig('1.jpg',bbox_inches='tight' )
plt.show()
```
